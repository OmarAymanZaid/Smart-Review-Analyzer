```python
data = data.drop(columns=[
    'Reviewer Name',
    'Profile Link',
    'Review Count',
    'Review Date',
    'Date of Experience'
])
```

```python
data = data.dropna(subset=['Review Text', 'Rating'])
data['Review Title'] = data['Review Title'].fillna('')
data['Country'] = data['Country'].fillna('Unknown')
```

```python
data['Rating'] = data['Rating'].str.extract(r'(\d)').astype(float)
data = data.dropna(subset=['Rating'])
data['Rating'] = data['Rating'].astype(int)
```

```python
data['Sentiment'] = data['Rating'].apply(lambda x: 1 if x >= 4 else 0 if x <= 2 else None)
data = data.dropna(subset=['Sentiment'])
data['Sentiment'] = data['Sentiment'].astype(int)
```

```python
data['Full Review'] = data['Review Title'] + " " + data['Review Text']
```

```python
data.to_csv("../data/processed/clean_reviews.csv", index=False)
```
